### üîπüìå Section 1: Loading Dataset and Initial Overview

In this step, we import the necessary Python libraries for data analysis and visualization, including `pandas`, `numpy`, `matplotlib`, `seaborn`, and `plotly.express`.  

The Netflix dataset (`netflix.csv`) is then loaded into a DataFrame using `pd.read_csv()`.  

`df.head()` displays the first five rows to provide an initial view of the dataset and its structure.


```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px


df = pd.read_csv("../data/raw/netflix.csv")

# Top rows
df.head()
```

## Remove Duplicate Records

- Displays the number of rows before removing duplicates  
- Drops duplicate rows from the dataset using `drop_duplicates()`  
- Displays the number of rows after removing duplicates

```python
print("Rows before dropping duplicates:", len(df))
df.drop_duplicates(inplace=True)
print("Rows after dropping duplicates:", len(df))
```

## üìå Section 2: Data Cleaning

- **Remove leading/trailing spaces** in string columns to ensure consistent entries  
- **Convert `date_added`** to datetime format for time-based analysis  
- **Fill missing values** in important categorical columns (`director`, `cast`, `country`, `rating`) with `"Unknown"`  
- **Split `duration`** into numeric (`duration_int`) and unit (`duration_unit`) for easier analysis  
- **Convert `type` column** to categorical type for memory efficiency and categorical operations  
- **Recheck dataset info and missing values** to confirm cleaning was successful  

This step ensures the dataset is consistent, complete, and ready for further exploration and feature engineering.


```python
# üìå Section 2 ‚Äî Data Cleaning

# 1Ô∏è‚É£ Remove leading/trailing spaces in string columns
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)

# 2Ô∏è‚É£ Convert 'date_added' to datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# 3Ô∏è‚É£ Fill missing values in important categorical columns
fill_cols = ['director', 'cast', 'country', 'rating']
df[fill_cols] = df[fill_cols].fillna("Unknown")

# 4Ô∏è‚É£ Split 'duration' into numeric + unit (min or seasons)
df[['duration_int', 'duration_unit']] = df['duration'].str.split(" ", expand=True)
df['duration_int'] = pd.to_numeric(df['duration_int'], errors='coerce')

# 5Ô∏è‚É£ Convert type column properly
df['type'] = df['type'].astype('category')

# 6Ô∏è‚É£ After cleaning ‚Äî recheck info and missing values
print("\nUpdated Data Info:")
display(df.info())

print("\nMissing Values After Cleaning:")
print(df.isnull().sum().sort_values(ascending=False))



```

##  Save Cleaned Dataset

- Creates a `processed` folder inside `data` if it does not exist  
- Saves the cleaned dataset as `cleaned_data.csv` without the index column  
- Prints a confirmation message once the file is successfully saved  

This step ensures that the cleaned and preprocessed data is stored for future analysis without repeating the cleaning process.


```python
import os

os.makedirs("../data/processed", exist_ok=True)
df.to_csv("../data/processed/cleaned_data.csv", index=False)



print("File saved successfully!")
```

## üìå Section 3: Feature Engineering

- **Convert columns to categorical type** (`rating`, `country`, `duration_unit`) for optimized memory usage and easier categorical analysis  
- **Create `duration_category`** by binning `duration_int` into Short (<60 min), Medium (60-120 min), and Long (>120 min)  
- **Expand genres** by splitting the `listed_in` column and creating a new `Genre` column for detailed genre-level analysis  
- **Check updated dataset info** to confirm the new features are added successfully  

This step enhances the dataset with meaningful features for deeper exploratory and analytical insights.


```python
# üìå Section 3 ‚Äî Feature Engineering

# 1Ô∏è‚É£ Convert important columns to 'category' type
cat_cols = ['rating', 'country', 'duration_unit']
for col in cat_cols:
    df[col] = df[col].astype('category')

# 2Ô∏è‚É£ New feature: Duration Category for Movies
df['duration_category'] = pd.cut(
    df['duration_int'],
    bins=[0, 60, 120, 300],
    labels=['Short (<60 min)', 'Medium (60-120 min)', 'Long (>120 min)']
)

df['duration_category'] = df['duration_category'].astype('category')

# 3Ô∏è‚É£ Expand genres for deeper analysis
df_genres = df.assign(Genre=df['listed_in'].str.split(', ')).explode('Genre')

# Show updated info
df.info()
```

### üìåSection 4: Data Visualization
üîπ Distribution of Movies vs TV Shows

We visualize the count of Movies and TV Shows available on Netflix using a count plot.

### üîπ Content Added to Netflix Over the Years

A new feature `year_added` is created from the `date_added` column to extract the year each title was added.  

We then visualize the distribution of content added per year using a count plot.  

This analysis highlights trends in Netflix's content expansion over time and provides insights into periods of increased content acquisition.


```python
df['year_added'] = df['date_added'].dt.year
df['year_added'] = df['year_added'].astype('Int64')

plt.figure(figsize=(12,6))
sns.countplot(x='year_added', data=df, order=sorted(df['year_added'].dropna().unique()))
plt.xticks(rotation=45)
plt.title("Content Added to Netflix Over the Years")
plt.savefig("../outputs/content_added_over_years.png")  # Correct path
plt.show()
```

### üîπ Distribution of Netflix Ratings

The distribution of content ratings on Netflix is visualized using a horizontal count plot.  

This analysis provides insights into the prevalence of different content ratings and helps understand Netflix's content classification trends.


```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, y='rating', order=df['rating'].value_counts().index)
plt.title("Distribution of Netflix Ratings")
plt.xlabel("Count")
plt.ylabel("Rating")
plt.savefig("../outputs/rating_distribution.png")
plt.show()
```

### üîπ Content Added by Year

The `year_added` feature is extracted from the `date_added` column to represent the year each title was added to Netflix.  

A line plot is then created to visualize the number of titles added per year.  

This trend analysis highlights periods of significant content addition and provides insights into Netflix's expansion over time.


```python
df['year_added'] = df['date_added'].dt.year
df['year_added'] = df['year_added'].astype('Int64')

titles_per_year = df['year_added'].dropna().value_counts().sort_index()

plt.figure(figsize=(12,6))
plt.plot(titles_per_year.index, titles_per_year.values, marker='o')
plt.title("Content Added by Year")
plt.xlabel("Year Added on Netflix")
plt.ylabel("Number of Titles")
plt.grid(True)
plt.savefig("../outputs/content_added_by_year.png")
plt.show()
```

### üîπ Rating Distribution by Content Type

### üîπ Trend of Movies & TV Shows Added Over the Years

We analyze the yearly addition of Movies and TV Shows by grouping the dataset by `year_added` and `type`.  

A line chart is created to visualize trends over time, highlighting patterns in content acquisition for different types.  

This analysis provides insights into Netflix's strategy in expanding its Movies and TV Shows catalog across the years.


```python
year_type_count = df.groupby(['year_added', 'type'], observed=False).size().reset_index(name='count')

fig = px.line(
    year_type_count,
    x='year_added',
    y='count',
    color='type',
    markers=True,
    title="Trend of Movies & TV Shows Added Over the Years"
)

fig.update_layout(
    xaxis_title="Year Added",
    yaxis_title="Number of Titles",
    template="plotly_white"
)
plt.savefig("../outputs/trend_movies_tv_shows.png")
fig.show()
```

### üîπ Heatmap of Titles Added per Year

A heatmap is generated to visualize the number of Movies and TV Shows added to Netflix each year.  

The dataset is pivoted by `year_added` and `type` to calculate counts, and the heatmap highlights patterns in content addition over time.  

This visualization helps identify years with high or low content additions and compare trends between Movies and TV Shows.


```python
import seaborn as sns
plt.figure(figsize=(14, 6))

heatmap_data = df.pivot_table(index='year_added', columns='type', aggfunc='size', fill_value=0,observed=False)

sns.heatmap(heatmap_data, cmap='YlOrRd', annot=False)
plt.title("Heatmap: Titles Added per Year (Movies vs TV Shows)")
plt.xlabel("Type")
plt.ylabel("Year Added")
plt.savefig("../outputs/heatmap_titles_added.png")
plt.show()
```
